Prelab

Context managers en Python: Es un objeto que define métodos especiales (__enter__ y __exit__) para gestionar recursos automáticamente, generalmente mediante la sentencia with. Su función principal es preparar un entorno antes de ejecutar un bloque de código y limpiarlo después (cerrar archivos, liberar memoria, liberar bloqueos), incluso si ocurren errores, evitando fugas de memoria y errores de bloqueo

Fuente: El libro de python

```python
with open("archivo.txt", "r", encoding="utf-8") as archivo:
    contenido = archivo.read()
    print(contenido)
```

DataFrames en Python:Un DataFrame en Python es una estructura de datos bidimensional, tabular y mutable de la biblioteca pandas, diseñada para organizar datos en filas y columnas. Funciona como una tabla SQL o una hoja de cálculo (Excel), ideal para análisis de datos, permitiendo almacenar tipos de datos heterogéneos y manipular grandes volúmenes de información de manera eficiente

Fuente: Microsoft Support

```python
import pandas as pd

datos = {
    "Nombre": ["Ana", "Luis", "Carlos"],
    "Edad": [20, 21, 19]
}

df = pd.DataFrame(datos)
print(df)
```

Archivos Pickle en Python:Implementa protocolos binarios para serializar y deserializar una estructura de objetos Python. Pickling es el proceso mediante el cual una jerarquía de objetos de Python se convierte en una secuencia de bytes, y el unpickling es la operación inversa, mediante la cual una secuencia de bytes de un archivo binario o un objeto tipo binario es convertido nuevamente en una jerarquía de objetos. Pickling y unpickling son alternativamente conocidos como serialización, ensamblaje o aplanamiento; sin embargo, para evitar confusiones, los términos utilizados aquí son pickling y unpickling.


Fuente: Python documentation

```python
import pickle

datos = {"nombre": "Ana", "edad": 20}

with open("datos.pkl", "wb") as archivo:
    pickle.dump(datos, archivo)
```

EJERCICIO 1:
En este ejercicio se trabajó con la lectura y escritura de archivos CSV y TXT utilizando context managers en Python. El objetivo fue comprender cómo abrir, leer, escribir y cerrar archivos de forma segura mediante el uso de `with open()`.

```python
#Escribir TXT
with open("archivo.txt", "w", encoding="utf-8") as archivo:
    archivo.write("Hola, este es un archivo de prueba en formato TXT.")
```

```python
#Leer TXT
with open("archivo.txt", "r", encoding="utf-8") as archivo:
    contenido = archivo.read()
    print(contenido)
```

```python
#Escribir CSV
import csv

with open("archivo.csv", "w", newline="", encoding="utf-8") as archivo:
    escritor = csv.writer(archivo)
    escritor.writerow(["Nombre", "Edad"])
    escritor.writerow(["Ana", 20])
    escritor.writerow(["Luis", 22])
```

```python
#Leer CSV
import csv

with open("archivo.csv", "r", encoding="utf-8") as archivo:
    lector = csv.reader(archivo)
    for fila in lector:
        print(fila)
```

Pregunta: ¿En qué parte o partes de un Pipeline de Datos se pueden usar las operaciones de Lectura (R) y Escritura (W)?

Respuesta:
Las operaciones de lectura se utilizan principalmente en la etapa de extracción de datos, cuando la información se obtiene desde archivos, bases de datos o APIs. Las operaciones de escritura se utilizan en la etapa de carga o almacenamiento, cuando los datos ya procesados se guardan en archivos, bases de datos o sistemas de destino. Además, ambas operaciones pueden utilizarse en etapas intermedias para guardar resultados temporales durante el procesamiento.

Pregunta: ¿Se podría usar un DataFrame para todos los tipos de datos (Estructurados, No Estructurados, Semi Estructurados)?

Respuesta: No, un DataFrame está diseñado principalmente para trabajar con datos estructurados, es decir, datos organizados en filas y columnas. También puede adaptarse en ciertos casos a datos semi estructurados, siempre que puedan transformarse a una forma tabular. Sin embargo, no es la mejor opción para datos no estructurados, como imágenes, audio o texto libre, ya que estos requieren otras herramientas y formatos más especializados.

```python
#Escribir Pickle
import pickle

datos = {
    "nombre": "Ana",
    "edad": 20,
    "carrera": "Ingeniería"
}

with open("datos.pkl", "wb") as archivo:
    pickle.dump(datos, archivo)
```

```python
#Leer Pickle
import pickle

with open("datos.pkl", "rb") as archivo:
    datos_cargados = pickle.load(archivo)

print(datos_cargados)
```

Pregunta: ¿Qué tipo de dato se podría almacenar con un Pickle?

Respuesta: Con Pickle se pueden almacenar muchos tipos de objetos de Python, como listas, diccionarios, tuplas, conjuntos, DataFrames e incluso modelos entrenados en algunos proyectos de análisis de datos o machine learning. Esto lo hace útil cuando se necesita guardar información compleja para usarla nuevamente más adelante.
