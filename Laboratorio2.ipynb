### 1. Análisis Exploratorio de Datos (EDA) con Pandas

Definición: El EDA (Exploratory Data Analysis) es el proceso de analizar y comprender un conjunto de datos antes de aplicar modelos o tomar decisiones.

Métodos principales en Pandas


*   head() → muestra las primeras
filas
*   tail() → muestra las últimas filas
*   info() → información general del dataset
*   describe() → estadísticas básicas
*   shape → dimensiones del dataset
*   columns → nombres de columnas


Limpieza de datos


*   Manejo de valores nulos: isnull(), dropna(), fillna()
*   Eliminación de duplicados: drop_duplicates()
*   Cambio de tipos de datos: astype()






Análisis básico



*   Filtrado de datos
*   Agrupaciones: groupby()
*   Conteo de valores: value_counts()
*   Correlaciones: corr()

### Ejercicio 2: Gráficos

2.1 Scatter plot

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Age", y="Fare")
plt.title("Relación entre Edad y Tarifa")
plt.show()
```

El gráfico de dispersión muestra la relación entre la edad y la tarifa pagada, permitiendo identificar posibles patrones o concentraciones.

2.2 Bar plot

```python
df["Sex"].value_counts().plot(kind="bar")
plt.title("Cantidad de pasajeros por sexo")
plt.show()
```

El gráfico de barras permite comparar la cantidad de pasajeros según su sexo.

2.3 Histograma

```python
df["Age"].hist(bins=20)
plt.title("Distribución de edades")
plt.show()
```

El histograma muestra la distribución de edades de los pasajeros, permitiendo identificar concentraciones y dispersión.

2.4 ¿Cómo elegir el gráfico?

El tipo de gráfico depende del tipo de variable y del objetivo del análisis. Los histogramas se utilizan para variables numéricas, los gráficos de barras para variables categóricas y los scatter plots para analizar relaciones entre variables numéricas.

### Ejercicio 3: Complementar análisis con gráficos

3.1 Cuantitativas

```python
sns.histplot(df["Fare"], bins=20)
plt.title("Distribución de tarifas")
plt.show()
```

```python
sns.scatterplot(data=df, x="Age", y="Fare")
plt.show()
```

Los gráficos permiten visualizar la distribución y relaciones entre variables numéricas, complementando el análisis estadístico.

3.2 Cualitativas

```python
df["Pclass"].value_counts().plot(kind="bar")
plt.title("Clases de pasajeros")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Proporción por sexo")
plt.ylabel("")
plt.show()
```

Los gráficos permiten representar frecuencias y proporciones de variables categóricas, facilitando su interpretación.
